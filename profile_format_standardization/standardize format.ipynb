```python
def read_file_content(file_name,path):
    with open(path+'/'+file_name, "rb") as file:
        content = file.read()
    return content
```

```python
def parse_date_range(date_range_str):
        if '' in date_range_str : 
            
            start_year, end_year = date_range_str.split('  ')
        else : 
            start_year = date_range_str 
            end_year = 'Aujourdhui'
        if 'Aujourdhui' in end_year.split() or 'present' in end_year.lower() :  # Replace 'Aujourdhui' with current date 
            end_date = datetime.now()
        else : 
            end_date = parse_date(end_year)
        start_date = parse_date(start_year)
        return start_date, end_date
```

```python
from datetime import datetime, timedelta
def calculate_total_experience(date_ranges):
    try : 
        min_start_date = datetime.now()
        max_end_date = datetime.min
        if len(date_ranges) > 0 : 
            for date_range_str in date_ranges:
                start_date, end_date = parse_date_range(date_range_str)
                if start_date and end_date:
                    if start_date < min_start_date:
                        min_start_date = start_date
                    if end_date > max_end_date:
                        max_end_date = end_date

            total_experience = max_end_date - min_start_date
            return total_experience.days
        return timedelta(days=0).days
    except : 
        print('an error has occured' )
```

```python
def get_experience(row) : 
    positions = row['positions'] 
    date_ranges = [elem['date1'] for elem in row['positions'] if 'date1' in elem.keys()] 
    experience = calculate_total_experience(date_ranges) 
    return experience 
```

```python
data['days_experience'] = data.apply(get_experience,axis=1)
```

```python
def positions_wrapper(row) : 
    positions = list()
    for i in range(1,11) : 
        positions.append({
            'title' : row['organization_title_'+str(i)],
            'company_name' : row['organization_'+str(i)],
            'date_1' : row['organization_start_'+str(i)],
            'date_2' : row['organization_end_'+str(i)],
            'description' : row['position_description_'+str(i)]
        })
    return positions 
data_latest['positions'] = data_latest.apply(positions_wrapper, axis=1)

    
```

```python
def educations_wrapper(row) : 
    educations = list()
    for i in range(1,4) : 
        educations.append({
            'title' : row['education_'+str(i)],
            'date_1' : row['education_start_'+str(i)],
            'date_2' : row['education_end_'+str(i)],
            'description' : row['education_description_'+str(i)],
            'degree' : row['education_degree_'+str(i)]
        })
    return educations
data_latest['educations'] = data_latest.apply(educations_wrapper, axis=1)

```

```python
def split_skills(skills_list):
    final_result = []
    try : 
        list_skills = skills_list.split(',')
        for elem in list_skills:
            if elem.strip() != 'null':  
                if '(' in elem and ')' in elem:
                    parts = elem.split('(')
                    main_skill = parts[0].strip()
                    final_result.append(main_skill)
                    for part in parts[1:]:
                        sub_skill = part.split(')')[0].strip()
                        if sub_skill:  
                            final_result.append(sub_skill)
                else:
                    elem = elem.split(':')[0].strip()
                    final_result.append(elem)
        return final_result
    except : 
        return []
```

```python
from datetime import datetime

def extract_datetime_from_string_2(date_str):
    try:
        date_str = str(date_str).strip()  # Remove leading and trailing spaces
        if date_str.endswith('.0'):
            date_str = date_str[:-2] + '.01'  # Replace '.0' with '.01'
        date_format = "%Y.%m"  # Corrected year format
        date_str_with_day = date_str
        return datetime.strptime(date_str_with_day, date_format)
    except ValueError as e:
        print(f"Error processing date string '{date_str}': {e}")
        return None
```

```python
from datetime import datetime, timedelta
def calculate_total_experience_2(date_ranges):
        min_start_date = datetime.now()
        max_end_date = datetime.min
        if len(date_ranges) > 0 : 
            for date_range_str in date_ranges:
                start_date = extract_datetime_from_string_2(date_range_str['start'])
                if not math.isnan(date_range_str['end']):
                    end_date = extract_datetime_from_string_2(date_range_str['end']) 
                else : 
                    end_date = datetime.now()
                if start_date and end_date:
                    if start_date < min_start_date:
                        min_start_date = start_date
                    if end_date > max_end_date:
                        max_end_date = end_date

            total_experience = max_end_date - min_start_date
            return total_experience.days
        return timedelta(days=0).days

```

```python
def get_experience_2(row) : 
    positions = row['positions'] 
    date_ranges = [{'start': elem['date_1'], 'end': elem['date_2']} for elem in positions if 'date_1' in elem.keys() and 'date_2' in elem.keys() and not math.isnan(elem['date_1'])]    
    experience = calculate_total_experience_2(date_ranges) 
    return experience 
```

```python
import re

def match_program(text, programs):
    matched_program = None
    if text : 
        for program in programs:
            pattern = re.compile(r'\b' + re.escape(program.strip()) + r'\b', re.IGNORECASE)
            if re.search(pattern, text):
                matched_program = program.strip()
                return matched_program
    return None
text = "I am studying for my ingnieur degree in data science."
programs = ['licence','Doctoral', 'cycle ingenieur', 'cycle ingenieur','ingnieur','ingenieur', 'ingenierie en', 'master', 'mastere', 'phd', 'doctorat', 'engineering', 'engineer', 'bachelor', 'B.E', 'post-graduate', 'pre-engineering', 'preparatoire', 'prepa']

```

```python
def get_levels(educs) : 
    final_levels = list() 
    for educ in educs  : 
        if 'degree' in educ.keys() : 
            matched_program = match_program(educ['degree'], programs)
            if matched_program:
                educ_date2 = educ.get('date2', None)  
                if not educ_date2 : 
                    if assoc_dic[matched_program] == 'Bac + 3' : 
                        real_level = 'Bac + 2' 
                    elif assoc_dic[matched_program] == 'Bac + 4' : 
                        real_level = 'Bac + 3'
                    elif assoc_dic[matched_program] == 'Bac + 5' : 
                        real_level = 'Bac + 4'  
                    else :
                        real_level = assoc_dic[matched_program]
                    final_levels.append(real_level) 
                else : 
                    final_levels.append(assoc_dic[matched_program]) 

    return final_levels 
```

```python
def extract_max_level(educs): 
    if not len(educs): 
        return None

    levels = get_levels(educs)
    if not len(levels) : 
        return None
    level_values = {
        'Bac + 2' : 2,
        'Bac + 3': 3,
        'Bac + 4': 4,
        'Bac + 5': 5,
        'Doctorat': 6  
    }
    max_level = max(levels, key=lambda x: level_values.get(x, 0))
    return max_level
```
